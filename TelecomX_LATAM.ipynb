```python
import requests
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/Edwardo-g/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
request = requests.get(url)
data = request.json()
df = pd.json_normalize(data)
df.head()
```

```python
#Contar los valores vacios
for col in df.select_dtypes(include="object").columns:
    vacios = (df[col].str.strip() == "").sum()
    print(f'Datos vacíos en {col}: {vacios}')
```

```python
df['account.Charges.Total'] = pd.to_numeric(df['account.Charges.Total'], errors='coerce')
df.info()
```

```python
df.isna().sum()
```

```python
#Contando cantidad de datos que son invalidos que comparten fila
((df["Churn"] == "") & (df["account.Charges.Total"].isnull())).sum()
```

```python
df.drop(df[(df["Churn"] == "")].index,axis= 0, inplace=True)
df.dropna(inplace=True)
df.info()
```

```python
df['Cuentas_Diarias'] = df['account.Charges.Monthly'] / 30.4167 #Utlizo esto en lugar de 30 por que es mas cercano al valor real que tendria despues de un año
```

```python
#Variables de Telefono
variables_categoricas = ['servicio_telefonico','servicio_internet', 'lineas_multiples', 'seguridad_online',
                        'respaldo_online', 'proteccion_dispositivo',
                         'soporte_tecnico', 'streaming_tv', 'streaming_peliculas',]
sns.set(rc={'figure.figsize':(25,20)})

for i, var in enumerate(variables_categoricas):

    plt.subplot(3,3,i+1)
    tabla = pd.crosstab(df[var], df['cancelacion'])
    tabla_pct = tabla.div(tabla.sum(axis=1), axis=0)
    ax.legend(
    labels=["No Cancelado", "Cancelado"],
    title="Estado",
    loc="upper right"
    )
    ax = plt.gca()
    if var == 'servicio_telefonico':
        ax.set_xticks([0,1])
    bottom = None
    for col in tabla_pct.columns:
        ax.bar(
            tabla_pct.index,
            tabla_pct[col],
            bottom=bottom,
            label=col
        )

        if bottom is None:
            bottom = tabla_pct[col]
        else:
            bottom += tabla_pct[col]

    ax.set_ylim(0,1)
    #vals = [0,0.25,0.5,0.75,1]
    ax.set_yticks(vals)
    ax.set_yticklabels([f"{v:.0%}" for v in vals])
    ax.set_ylim(0,1.3)

    ax.set_title(f'Tasa de cancelación - {var}', fontsize=15)
    ax.set_ylabel('Porcentaje', fontsize=12)
    ax.set_xlabel(var, fontsize=12)

    for j, categoria in enumerate(tabla_pct.index):
        porcentaje = tabla_pct.loc[categoria, 1]

        ax.text(
            j,
            1.02,
            f"{porcentaje:.1%}",
            ha='center',
            fontweight='bold'
        )

plt.tight_layout()
plt.show()
```

La evasión de clientes, conocida como churn, representa uno de los principales desafíos en empresas de servicios, especialmente en el sector de telecomunicaciones. La pérdida constante de clientes impacta directamente en los ingresos y en la estabilidad del negocio, además de incrementar los costos asociados a la adquisición de nuevos usuarios.

El objetivo de este análisis es identificar patrones y variables que puedan estar influyendo en la cancelación del servicio, con el fin de comprender mejor el comportamiento de los clientes y detectar posibles factores de riesgo. A través de un análisis exploratorio de datos, se examinarán variables como la antigüedad del cliente, los cargos mensuales y otras características demográficas y contractuales.

En una primera revisión descriptiva, observamos una tasa de cancelación cercana al 27%, lo cual representa un nivel considerable y potencialmente preocupante para la empresa. Este hallazgo inicial justifica profundizar en el análisis para determinar qué características diferencian a los clientes que cancelan del resto.

Asimismo, variables como la antigüedad en meses muestran una alta dispersión, lo que sugiere que el tiempo de permanencia podría jugar un papel clave en el comportamiento de cancelación. De igual manera, los cargos mensuales presentan variabilidad significativa, lo que podría estar relacionado con distintos niveles de sensibilidad al precio.

## **Limpieza y Tratamiento de Datos**

Antes de iniciar el análisis exploratorio, fue necesario realizar un proceso de limpieza y transformación de los datos con el objetivo de garantizar su consistencia, calidad y correcta interpretación.

En primer lugar, los datos originales presentaban estructuras en formato de diccionarios anidados, propias del formato de origen. Estas estructuras fueron normalizadas y desanidadas, permitiendo convertir la información en un formato tabular plano (dataframe), adecuado para su manipulación y análisis posterior.

Posteriormente, se realizó una revisión general para identificar valores inválidos, inconsistencias y datos faltantes. Las filas que contenían información incompleta o errónea fueron eliminadas, ya que podrían introducir sesgos o afectar la precisión de los análisis estadísticos y visualizaciones. Esta depuración permitió trabajar con un conjunto de datos más limpio y confiable.

Una vez depurados los registros, se procedió a ajustar los tipos de datos de cada columna, asegurando que las variables numéricas, categóricas y binarias estuvieran correctamente definidas. Este paso es fundamental para evitar errores en cálculos posteriores y garantizar que las funciones estadísticas se apliquen correctamente.

Además, las variables con respuestas binarias expresadas como “Sí” o “No” fueron transformadas a valores numéricos (1 y 0, respectivamente). Esta conversión facilita el cálculo de proporciones, correlaciones y la construcción de modelos predictivos, así como la generación de gráficos más eficientes.

Finalmente, se realizó una estandarización en los nombres de las columnas, reemplazando etiquetas originales por nombres más claros, consistentes y fáciles de interpretar. Esto mejora la legibilidad del código y simplifica el proceso de análisis.

En conjunto, estas etapas de limpieza y tratamiento permitieron obtener un dataset estructurado, coherente y listo para el análisis exploratorio y la posterior modelación del problema de evasión de clientes (churn).

En este caso, podemos observar que, aunque el electronic check presenta un porcentaje significativamente mayor de cancelaciones —aproximadamente un 45% de los clientes que lo utilizan terminan cancelando el servicio—, eliminar este método de pago podría ser una decisión equivocada.

En primer lugar, su eliminación podría colocarnos en desventaja frente a la competencia, especialmente si otras empresas continúan ofreciendo esta alternativa. Además, se trata del método de pago más utilizado por nuestros clientes, representando cerca del 33.6% del total y alrededor del 25% de los clientes que permanecen activos. Esto indica que, a pesar de su alta tasa de cancelación, sigue siendo un canal relevante dentro de nuestra base de usuarios.

Por lo tanto, en lugar de eliminarlo, sería más prudente investigar las posibles causas de insatisfacción asociadas a este método de pago. La implementación de encuestas o estudios de experiencia del cliente podría ayudarnos a identificar fricciones específicas —como problemas en el proceso de pago, cargos inesperados o dificultades técnicas— y, a partir de ello, desarrollar estrategias de mejora.

En conclusión, el problema no necesariamente radica en la existencia del método de pago, sino en la experiencia que los clientes tienen al utilizarlo. Mejorarlo podría reducir significativamente la tasa de cancelación sin sacrificar una opción ampliamente demandada.

Más allá de posibles estrategias como ofrecer descuentos iniciales o planes promocionales para incentivar la permanencia durante los primeros meses —una etapa especialmente vulnerable—, estos datos no señalan una acción inmediata y específica. Sin embargo, sí proporcionan un contexto fundamental para la toma de decisiones estratégicas y el diseño de futuras iniciativas de retención.

```python
#Variables de Telefono
variables_categoricas = ['servicio_telefonico','adulto_mayor','servicio_internet',
                         'seguridad_online','respaldo_online', 'proteccion_dispositivo',
                         'soporte_tecnico', 'streaming_tv', 'streaming_peliculas']
sns.set(rc={'figure.figsize':(25,20)})

for i, var in enumerate(variables_categoricas):

    plt.subplot(3,3,i+1)
    tabla = pd.crosstab(df[var], df['cancelacion'])
    tabla_pct = tabla.div(tabla.sum(axis=1), axis=0)
    ax.legend(
    labels=["No Cancelado", "Cancelado"],
    title="Estado",
    loc="upper right"
    )
    ax = plt.gca()
    if var == 'servicio_telefonico':
        ax.set_xticks([0,1], ['No', 'Si'])
    bottom = None
    for col in tabla_pct.columns:
        ax.bar(
            tabla_pct.index,
            tabla_pct[col],
            bottom=bottom,
            label=col
        )

        if bottom is None:
            bottom = tabla_pct[col]
        else:
            bottom += tabla_pct[col]
    vals = [0.2, 0.4, 0.6, 0.8, 1]
    ax.set_yticks(vals)
    ax.set_ylim(0,1)
    #vals = [0,0.25,0.5,0.75,1]
    ax.set_yticks(vals)
    ax.set_yticklabels([f"{v:.0%}" for v in vals])
    ax.set_ylim(0,1.3)

    ax.set_title(f'Tasa de cancelación - {var}', fontsize=15)
    ax.set_ylabel('Porcentaje', fontsize=12)
    ax.set_xlabel(var, fontsize=12)

    for j, categoria in enumerate(tabla_pct.index):
        porcentaje = tabla_pct.loc[categoria, 1]

        ax.text(
            j,
            1.02,
            f"{porcentaje:.1%}",
            ha='center',
            fontweight='bold'
        )

plt.tight_layout()
plt.show()
```

En primer lugar, el servicio de línea telefónica no parece tener una relación significativa con las cancelaciones. La diferencia en la tasa de abandono entre quienes lo tienen y quienes no es inferior al 2%. Además, el ligero incremento observado puede estar más relacionado con el aumento en el precio total del paquete que con el servicio en sí.

Donde sí debemos prestar especial atención es en el servicio de internet. La tasa de cancelación es de apenas 7.4% entre quienes no lo contratan, aumenta a 19% en quienes utilizan DSL y se eleva de forma preocupante a 41.9% en los clientes con fibra (o el plan de mayor velocidad). Si bien parte de este incremento podría explicarse por el precio —con un costo promedio de 91.5 dólares frente a los 58.09 dólares del DSL—, la diferencia de más del doble en la tasa de cancelación sugiere que el problema no se debe únicamente al factor económico. Es necesario investigar qué debilidades o insatisfacciones están percibiendo estos clientes y diseñar un plan de mejora específico.

Otro punto importante es el 41.7% de cancelación en personas mayores. Esta cifra es muy similar a la observada en clientes que no contratan servicios adicionales como soporte técnico (41.8%), seguridad online (41.6%), respaldo en línea (39.9%) y protección de dispositivos (39.1%). En contraste, las tasas de cancelación disminuyen considerablemente entre quienes sí cuentan con estos servicios: soporte técnico (15.2%), seguridad online (14.6%), respaldo en línea (21.6%) y protección de dispositivos (22.5%). Esto evidencia la importancia de estos complementos y sugiere que podría ser prudente ofrecerlos por defecto —salvo que el cliente indique lo contrario—, especialmente en el segmento de personas mayores donde podemos notar un 50% de abandonos para los que no contrataron el soporte tecnico, queda claro que este segmento es especialmente sensible a la ausencia de apoyo.

Por último, aunque el servicio de streaming muestra una ligera mejora en la retención, la diferencia es moderada: 33% de cancelación en quienes no lo tienen frente a 30% en quienes sí lo contratan. Esto indica que, aunque aporta valor, no es un factor determinante en comparación con otras variables que claramente requieren mayor atención estratégica.

##**Conclusiones**

Uno de los focos principales es el método de pago Electronic Check, que presenta una tasa de abandono cercana al 34%, significativamente superior a otras modalidades. Este comportamiento lo convierte en uno de los factores más relevantes dentro del análisis de cancelación.

Asimismo, el servicio de fibra óptica muestra una tasa de cancelación de 41.9%, más del doble que su contraparte DSL (19%). Esta diferencia marca una brecha considerable entre ambos tipos de servicio de internet.

Otro hallazgo especialmente significativo corresponde al segmento de personas mayores. Este grupo, que representa aproximadamente el 37% de la base de clientes, presenta una tasa de abandono del 50% cuando no cuenta con soporte técnico. Incluso en clientes más jóvenes, la ausencia de soporte eleva la cancelación a casi 39%. En contraste, cuando el servicio de soporte está activo, la tasa se reduce a 19.6% en adultos mayores y a 14.7% en personas más jóvenes, lo que evidencia una diferencia sustancial en retención según la disponibilidad de asistencia.

Finalmente, se observa una retención particularmente baja durante los primeros meses de contratación, lo que confirma que esta etapa representa uno de los períodos de mayor vulnerabilidad en el ciclo de vida del cliente.

En conjunto, los resultados sugieren que la cancelación de clientes está fuertemente asociada a factores como el método de pago, el tipo de servicio de internet, la disponibilidad de soporte técnico y los primeros meses de relación con la empresa.
